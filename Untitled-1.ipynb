```python
col = [2016,2017,2018,2019,2020,2021,2022]
index = [1,2,3,4,5,6,7,8]
mat = pd.DataFrame(0, index = index, columns = index) #Creation of the empty transition matrix
for i in df["Id"].unique(): #iterating when the Id
	l = df[df["Id"] == i].reset_index() #Considering when Id are the same ( to get only one company)
	if len(l) > 1: # if < 1 ==> no transition happened
		for j in range(1, len(l) - 1):
			if l.loc[j]["Rating"] != l.loc[j+1]["Rating"] and l.loc[j]["Date"].year != l.loc[j+1]["Date"].year: #not taking into account transition within a year only from year n to n+1
				mat.loc[l.loc[j]["Rating"]][l.loc[j+1]["Rating"]] += 1

		for d in l["Date"].unique(): #Transition within a year
			l_2 = l[l["Date"] == np.datetime_as_string(d, unit='Y')[:4]].reset_index() # considering when the year are the same
			if len(l_2) > 1 and l_2.loc[0]["Rating"] != l_2.loc[-1]["Rating"]: #considering when there is more than one transition within a year and considering only the first and and the last transition
				mat.loc[l.loc[j]["Rating"]][l.loc[j+1]["Rating"]] += 1 # Add one when a tansition whithin a year is found
	else : 
		mat.loc[l.loc[0]["Rating"]][l.loc[0]["Rating"]] += 1
```

```python
trans_mat = mat / mat.sum(axis=0) #should the sum ba one on the axis 0 or 1
```
