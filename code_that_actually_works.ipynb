```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("transition_data.xlsx", date_parser='-')
#Creation of a dataframe only with the year
df['Year'] = df['Date'].dt.year

new_df = df.drop_duplicates(subset=['Id','Year'], keep='last')

index = [1,2,3,4,5,6,7,8]
mat = pd.DataFrame(0, index = index, columns = index) #Creation of the empty transition matrix
for i in new_df["Id"].unique(): #iterating when the Id
	l = new_df[new_df["Id"] == i].reset_index() #Considering when Id are the same ( to get only one company)
	if len(l) > 1:
		for j in range(len(l)-1):
			mat.loc[l.loc[j]["Rating"]][l.loc[j+1]["Rating"]] += 1
	else : 
		mat.loc[l.loc[0]["Rating"]][l.loc[0]["Rating"]] += 1  
```

```python
mat

trans_mat = mat / mat.sum(axis=0) #should the sum ba one on the axis 0 or 1
```
